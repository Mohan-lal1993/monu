```python
import pandas as pd
```

```python
nba=pd.read_csv("nba.csv")
```

```python
nba.sort_values(["Number", "Salary", "Name"], inplace=True)
```

```python
nba=nba.sort_index(ascending=True)
```

```python
nba
```

# Rank Value with rank() Method

```python
nba = nba.dropna(how="all")
```

```python
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
```

```python
nba["Salary"].value_counts()
```

```python
nba["Salary"].rank()
```

```python
nba.sort_values(by="Salary")
```

```python
nba.sort_values(by="Salary",ascending=False)
```
