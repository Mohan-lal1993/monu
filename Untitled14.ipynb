# Retrieve Row by Index Label with .loc[]

```python
import pandas as pd
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
```

```python
bond.sort_index(inplace=True)
```

```python
bond
```

```python
bond.loc["Goldfinger"]
```

```python
bond.loc["Diamonds are forever":"Goldfinger"]
```

```python
bond.loc[:"Goldfinger"]
```

```python
bond.loc[["Moonraker","Goldfinger"]]
```

```python
bond.loc[["Goldfinger","gold"]]
```

```python
"gold" in bond.index
```
